```python
# 🌧️ Select all months from all available years
ds_month = ds_JPC.where(ds_JPC.time.dt.month == 12, drop=True)
```

```python
# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot
sc = ax.scatter(
    ds_month.d_m_filt,
    np.log10(ds_month.n_w_filt),
    c=ds_month.lwc_filt,
    cmap='viridis',
    norm=mcolors.Normalize(vmin=0, vmax=13),
    s=30,
    edgecolors='none'
)

# --- Maritime rectangle ---
rect1 = patches.Rectangle(
    (1.5, 4),
    width=0.25,
    height=0.5,
    linewidth=2,
    edgecolor='red',
    facecolor='none'
)

# --- Continental rectangle ---
rect2 = patches.Rectangle(
    (2, 3),
    width=0.75,
    height=0.5,
    linewidth=2,
    edgecolor='red',
    facecolor='none'
)

ax.add_patch(rect1)
ax.add_patch(rect2)

# Labels
ax.set_xlabel('$D_m\ [mm]$', fontsize=14)
ax.set_ylabel('$\log_{10}N_W\ [mm^{-1}\ m^{-3}]$', fontsize=14)

# Title
# ax.set_title(f'February Climatology ({start_date} to {end_date})', fontsize=16)
ax.set_title(f'December Climatology', fontsize=16)

# Text labels
plt.text(1.4, 3.6, 'Continental', fontsize=14, color='red')
plt.text(1.75, 4.6, 'Maritime', fontsize=14, color='red')

# Colorbar
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('LWC $[g\ m^{-3}]$', fontsize=14)

# Minor ticks
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())

plt.tight_layout()
plt.savefig(r'C:\Users\juanp\Documents\Universidad\Trabajo_de_Grado_Isabel\Articulo_Trabajo_Grado\Imagenes\Torre_SIATA\Torre_SIATA_Decembers.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 📂 Folder containing all monthly figures for ONE instrument
ruta_imagenes = r"C:\Users\juanp\Documents\Universidad\Trabajo_de_Grado_Isabel\Articulo_Trabajo_Grado\Imagenes\Torre_SIATA"

# 📅 Define the position of each month in the matrix
# (row, column)
month_positions = {
    "Januarys":   (0, 1),
    "Februarys":  (0, 2),
    "Marchs":     (1, 0),
    "Aprils":     (1, 1),

    "Mays":       (1, 2),
    "Junes":      (2, 0),
    "Julys":      (2, 1),
    "Augusts":    (2, 2),

    "Septembers": (3, 0),
    "Octobers":   (3, 1),
    "Novembers":  (3, 2),
    "Decembers":  (0, 0),

    "DEFs":      (0, 3),
    "MAMs":      (1, 3),
    "JJAs":      (2, 3),
    "SONs":      (3, 3)
}

# 🖼️ Create empty matrix figure
fig, axs = plt.subplots(4, 4, figsize=(20, 15))

# 🔁 Read all png files inside folder
files = glob.glob(os.path.join(ruta_imagenes, "*.png"))

# 🔍 Loop through images
for file in files:

    filename = os.path.basename(file)

    # 📅 Identify month from filename
    for month, (row, col) in month_positions.items():

        # Example filename:
        # Torre_SIATA_January.png
        if month in filename:

            # 📖 Read image
            img = mpimg.imread(file)

            # 📌 Plot image in corresponding position
            axs[row, col].imshow(img)

            # 🏷️ Optional title
            axs[row, col].set_title(month, fontsize=16)

            # ❌ Remove axes
            axs[row, col].axis('off')

# 🔧 Adjust spacing
plt.tight_layout()

# 💾 Save final matrix
plt.savefig(
    os.path.join(ruta_imagenes, "Torre_SIATA_Matrix.png"),
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
